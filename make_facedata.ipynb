```python
import numpy as np
import pandas as pd
import math
from scipy.spatial.transform import Rotation
import imageio
import os
import glob
```

```python
DIR = 'RAVDESS/processed'
files = glob.glob(os.path.join(DIR,'*.csv'))
```

```python
for file in files:
    print('Processing ', file, ' ', end='')
    name = os.path.splitext(os.path.basename(file))[0]
    
    if os.path.exists(os.path.join(DIR,name)+'.npz'):
        print('file exists. skip.')
        continue
    
    data = pd.read_csv(file, sep=',+\s+')

    d = []
    d_norm = []

    for f in range(len(data)):
        dd = []

        tx = data['pose_Tx'][f]
        ty = data['pose_Ty'][f]
        tz = data['pose_Tz'][f]
        rx = data['pose_Rx'][f]
        ry = data['pose_Ry'][f]
        rz = data['pose_Rz'][f]

        dd = [tx,ty,tz,rx,ry,rz]
        R = Rotation.from_euler('XYZ', [rx,ry,rz])

        for n in range(68):
            X = R.inv().apply(np.array([data['X_%d'%(n)][f]-tx,data['Y_%d'%(n)][f]-ty,data['Z_%d'%(n)][f]-tz]))
            dd.append(X[0])
            dd.append(X[1])        

        d.append(dd)

    d = np.array(d).T
    np.savez(os.path.join(DIR,name)+'.npz',d)
        
    d_norm = np.zeros(d.shape)
    for i in range(d.shape[0]):
        d_norm[i] = (d[i] - np.mean(d[i]))/np.std(d[i])

    imageio.imwrite(os.path.join(DIR,name)+'.png', d_norm)
```

```python
# Compute mean and std
DIR = 'RAVDESS/processed'
files = glob.glob(os.path.join(DIR,'*.npz'))

data_ALL = []

for file in files:
    print('Processing ', file)

    data = np.load(file)['arr_0']
    data_ALL.append(data)

data_ALL = np.concatenate(data_ALL,axis=1)

mean = np.mean(data_ALL,axis=1)
std = np.std(data_ALL,axis=1)

np.savez(os.path.join(DIR,'mean_std.npz'), mean=mean, std=std)
```
